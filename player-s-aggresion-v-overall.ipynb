```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress

```

```python
df = pd.read_csv('/kaggle/input/fc-24-players-database-and-stats-from-easports/all_players.csv')
```

# Correlation between Aggression and Overall Performance of the top 100 players 

```python
players_data = df.head(100)

x = players_data['Aggression']
y = players_data['Overall']

plt.figure(figsize=(10, 4))
sns.scatterplot(data=players_data, x='Aggression', y='Overall')

slope, intercept, r_value, p_value, std_err = linregress(x, y)

line = slope * x + intercept

plt.plot(x, line, 'r')
plt.title('Scatterplot of the 100 best players with Overall as a function of Aggression')


plt.legend()
plt.show()
```

# Top 10 most aggressive Clubs in football

```python
clubs = pd.DataFrame(df.groupby('Club')['Aggression'].sum().reset_index())
clubs = clubs.sort_values('Aggression', ascending=False).head(10).reset_index(drop=True)
clubs['Aggression'] = round(clubs['Aggression'], 2)

plt.figure(figsize=(8, 4))  # Adjust the figure size as needed

# Create the barplot and rotate x-axis labels
ax = sns.barplot(data=clubs, x='Club', y='Aggression')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

plt.show()
```

```python
#extracting male players from all players  
male_players = df[df['Gender']=='M']
male_players
```

```python
#extracting female players from all players 
female_players = df[df['Gender']=='F']
female_players
```

<h1>  Most aggressive male players by position </h1>

```python
top_aggressive_male_players_by_position = male_players.groupby('Position').apply(lambda group: group.nlargest(1, 'Aggression')).reset_index(drop=True)
male_players = top_aggressive_male_players_by_position[['Name','Position','Aggression', 'Overall']].sort_values(by='Aggression', ascending=False).reset_index(drop=True)
male_players 
```

```python
#correlation between male player's agreesion and overall performance 
male_correlation = male_players['Aggression'].corr(male_players['Overall'])
male_correlation
```

```python
top_aggressive_female_players_by_position = female_players.groupby('Position').apply(lambda group: group.nlargest(1, 'Aggression')).reset_index(drop=True)
female_players = top_aggressive_female_players_by_position[['Name','Position','Aggression', 'Overall']].sort_values(by='Aggression', ascending=False).reset_index(drop=True)
female_players
```

```python
#correlation between female player's agreesion and overall performance
female_correlation = female_players['Aggression'].corr(female_players['Overall'])
female_correlation
```

# Most aggressive male players by positions in football

```python
most_aggressive_male_players_positions = pd.DataFrame(male_players.groupby('Position')['Aggression'].sum('Aggression').reset_index())

plt.figure(figsize=(6, 4))
sns.barplot(data=most_aggressive_male_players_positions, x='Position', y= 'Aggression')
```

# Most aggressive female players by positions in football

```python
most_aggressive_female_players_positions = pd.DataFrame(female_players.groupby('Position')['Aggression'].sum('Aggression').reset_index())

plt.figure(figsize=(6, 4))
sns.barplot(data=most_aggressive_female_players_positions, x='Position', y= 'Aggression')
```

# I grouped the players into male and female player to investigate if the correction between aggression and overall applies to both male and female players  

Male Players (-0.3369): The negative correlation (-0.3369) between "Aggression" and "Overall" suggests that for male players, as "Aggression" increases, "Overall" tends to decrease. In other words, male players who are more aggressive may have lower overall ratings.

Female Players (0.7946): The positive correlation (0.7946) between "Aggression" and "Overall" for female players suggests that as "Aggression" increases, "Overall" also tends to increase. In this case, female players who are more aggressive may have higher overall ratings.

The difference in the sign and magnitude of the correlation coefficients indicates that the relationship between "Aggression" and "Overall" varies between male and female players. This could be due to differences in how aggression is perceived or valued in the game for male and female players. These findings may be of interest for coaches and analysts when assessing player performance and attributes based on gender-specific patterns.
